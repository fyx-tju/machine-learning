```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("../data/titanic/train.csv")
dataset.describe()
```

```python
dataset.head()
```

```python
dataset.Sex = dataset.Sex.map(lambda x: 0 if x=='male' else 1)
dataset.Cabin = dataset.Cabin.dropna()
dataset.head()
```

```python
#删除有缺失值的数据
from sklearn.model_selection import train_test_split

dataset = dataset.dropna(axis=0)
y = dataset.Age
feature = ["PassengerId","Pclass","Fare","Parch","SibSp"]
X = dataset[feature]
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
X.describe()
```

```python
#使用scikit-learn构建模型并训练
from sklearn.tree import DecisionTreeRegressor
#机器模型中会存在一些随机的过程，指定random_state可以确定每次得到相同的结果
decision_model = DecisionTreeRegressor(random_state=1)
decision_model.fit(X, y)
```

```python
#使用模型进行预测
from sklearn.metrics import mean_absolute_error
test_data = pd.read_csv("../data/titanic/test.csv")
test_data = test_data.dropna(axis=0)
test_x = test_data[feature]
test_y = test_data.Age
predict_value = decision_model.predict(test_x)
#print(predict_value)
#print(test_y)
print(mean_absolute_error(test_y,predict_value))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
get_mae(10,X,test_x,y,test_y)
```

```python
#随机森林
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(X, y)
melb_preds = forest_model.predict(test_x)
print(mean_absolute_error(test_y, melb_preds))
```
